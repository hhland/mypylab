```python
%%octave
x = 0:0.05:10;
y = sin(x) .* exp(-0.4*x);
subplot(2,2,1)
plot(x,y)
subplot(2,2,2)
ezplot('sin(x) * exp(-0.4*x)', [0,10])
subplot(2,2,3)
fplot('sin(x) * exp(-0.4*x)', [0,10])
```

```python
%%octave
t = 0:0.1:2*pi;
x = 4 * cos(t) - cos(4*t);
y = 4 * sin(t) - sin(4*t);
subplot(2,1,1)
plot(x,y)
subplot(2,1,2)
 ezplot('4*cos(t) - cos(4*t)', '4*sin(t) - sin(4*t)', [0, 2*pi])
```

```python
%%octave
subplot(2,1,1)
fplot('[sin(x), sin(2*x), sin(3*x)]', [0, 2*pi])
subplot(2,1,2)
fplot('[sin(x), sin(2*x), sin(3*x)]', [0, 2*pi],"*")
```

```python
%%octave 
t = 0:0.1:2*pi;
r = sin(2*t).* cos(2*t);
subplot(2,1,1)
polar(t,r)
subplot(2,1,2)
polar(t,r,"--r")
```

```python
%%octave
f = ['3 *(1-x)^2*exp(-(x^2)-(y + 1)^2)',...
'-10*(x/5-x^3-y^5)*exp(-x^2-y^2)','-1/3*exp(-(x + 1)^2 - y^2)'];
subplot(2,1,1)
ezcontour(f,[-3,3],49)
subplot(2,1,2)
ezcontourf(f,[-3,3],49)
```
