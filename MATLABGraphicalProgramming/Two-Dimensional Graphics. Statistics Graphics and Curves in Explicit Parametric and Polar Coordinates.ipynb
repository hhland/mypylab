```python
%%octave
x = linspace (0,2,30);
y = sin(x.^2);
plot(x,y)
text (1,0.8, ' y = sin(x^2)')
hold on
z = log(sqrt(x));
plot (x, z)
text (1, - 0.1, ' y = log (sqrt (x))')
xlabel('X Axis');
ylabel ('Y Axis');
title('Graphic sine and logarithmic');
```

```python
%%octave 
x = (0:0.1:2*pi);
y = sin(x);
z = cos(x);
subplot(1,2,1);
plot(x,y), axis([0, 2*pi, -1, 1]), title('sin(x)')
subplot(1,2,2);
plot(x,z), axis([0, 2*pi, -1, 1]), title('cos(x)')
```

```python
%%octave
x = (0:0.1:4*pi);
y = sin (x);
z = cos (x);
subplot(2,1,1);
plot(x,y), axis([0 2*pi -1 1]), title('sin(x)'), grid
subplot(2,1,2);
plot (x, z), axis([0 2*pi -1 1]), title ('cos (x)'), grid
```

```python
%%octave
 x = 0:0.01:3;
 y = abs(exp(-0.5*x) .* sin(5*x));
 subplot(2,2,1)
 plot(x,y)
 title('normal')
 hold on
 subplot(2,2,2)
 loglog(x,y)
 title('logarithmic')
 subplot(2,2,3)
 semilogx(x,y)
 title('semilogarithmic in X axis')
 subplot(2,2,4)
 semilogy(x,y)
 title('semilogarithmic in Y axis')
```
