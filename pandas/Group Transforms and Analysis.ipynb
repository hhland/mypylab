```python
%pylab inline
import pandas as pd
from numpy.random import rand
from pandas import DataFrame,Series
import string
import random
```

```python
np.random.seed(0)
N,M=1000,500

def rands(n):
    return ''.join([random.choice(string.ascii_uppercase) for _ in xrange(n)])

tickers=np.array([rands(5) for _ in xrange(N)])

tickers
```

```python
df=DataFrame({
    "Momentum":np.random.randn(M)/200+0.03,
    "Value":np.random.randn(M)/200+0.08,
    "ShortInterest":np.random.randn(M)/200-0.02,    
},index=tickers[:M])

df
```

```python
ind_name=np.array(["FINANCIAL","TECH"])
sampler=np.random.randint(0,len(ind_name),N)
industries=Series(ind_name[sampler],index=tickers,name="industry")

industries
```

```python
by_industry=df.groupby(industries)
print by_industry.mean()
print by_industry.describe()
```

```python
def zscore(group):
    return (group-group.mean())/group.std()

df_stand=by_industry.apply(zscore)

df_stand
```

```python
df_stand.groupby(industries).agg(["mean","std"])
```

```python
ind_rank=by_industry.rank(ascending=False)

ind_rank.groupby(industries).agg(["min","max"])
```

```python
by_industry.apply(lambda x: zscore(x.rank()))
```

```python
#Group Factor Exposures

fac1,fac2,fac3=rand(3,1000)
ticker_subset=tickers.take(np.random.permutation(N)[:1000])

print ticker_subset

```

```python
port=Series(0.7*fac1-1.2*fac2+.3*fac3+rand(1000),index=ticker_subset)

print port

factors=DataFrame({
    "f1":fac1,"f2":fac2,"f3":fac3
},index=ticker_subset)

factors
```

```python
factors.corrwith(port)
```

```python
pd.ols(y=port,x=factors).beta
```

```python
def beta_exposure(chunk,factors=None):
    return pd.ols(y=chunk,x=factors).beta

by_ind=port.groupby(industries)

exposures=by_ind.apply(beta_exposure,factors=factors)

exposures.unstack()
```

```python
#Decile and Quartile Analysis

import pandas.io.data as web
```

```python
data=web.get_data_yahoo("SPY","2006-01-01")
data
```

```python
px=data["Adj Close"]
returns=px.pct_change()

returns
```

```python
def to_index(rets):
    index=(1+rets).cumprod()

    first_loc=max(index.notnull().argmax()-1,0)
    index.values[first_loc]=1
    return index

def trend_signal(rets,lookback,lag):
    signal=pd.rolling_sum(rets,lookback,min_periods=lookback-5)
    return signal.shift(lag)
```

```python
signal=trend_signal(returns,100,3)
print signal
trade_friday=signal.resample("W-FRI").resample("B",fill_method="ffill")
print trade_friday
trade_rets=trade_friday.shift(1)*returns
print trade_rets
```

```python
to_index(trade_rets).plot()
```
