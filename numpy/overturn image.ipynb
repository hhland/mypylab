```python
%pylab inline
import scipy.misc
```

```python
lena=scipy.misc.lena()
plt.subplot(221)
plt.imshow(lena)
```

```python
plt.subplot(222)
lean_turnover=lena[:,::-1]
plt.imshow(lean_turnover)
```

```python
plt.subplot(223)
lena_lefttop=lena[:lena.shape[0]/2,:lena.shape[1]/2]
plt.imshow(lena_lefttop)
```

```python
plt.subplot(224)
lena_mash=lena.copy()
lena_mash[lena_mash%2==0]=0
plt.imshow(lena_mash)
plt.show()
```
