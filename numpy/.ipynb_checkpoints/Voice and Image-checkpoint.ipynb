```python
%pylab inline
import scipy
import requests
import scipy.io.wavfile as wavfile
import scipy.ndimage
from random import choice
import scipy.signal
import urllib2
```

```python
"""
load image into memory
"""
N,NSQUARES=512,30

img=zeros((N,N),np.uint8)
centers=np.random.random_integers(0,N,size=(NSQUARES,2))
print centers
radii=np.random.randint(0,N/9,size=NSQUARES)
print radii
colors=np.random.randint(100,255,size=NSQUARES)
print colors


```

```python
for i in xrange(NSQUARES):
    xinds=range(centers[i][0]-radii[i],centers[i][0]+radii[i])
    xinds=np.clip(xinds,0,N-1)
    yinds=range(centers[i][1]-radii[i],centers[i][1]+radii[i])
    yinds=np.clip(yinds,0,N-1)
    
    if len(xinds)==0 or len(yinds)==0:
        continue
    
    coor=np.meshgrid(xinds,yinds)
    img[coor]=colors[i]

fname="random_squares.raw"
img.tofile(fname)
img_memmap=np.memmap(fname,shape=img.shape)

plt.imshow(img_memmap)
plt.axis("off")
```

```python
from scipy import misc
""" mix image """
ITERATION=50
lean=misc.lena()
SIZE=lean.shape[0]
MAX_COLOR=255
x_min,x_max=-2.5,1
y_min,y_max=-1,1

x,y=np.meshgrid(np.linspace(x_min,x_max,SIZE),np.linspace(y_min,y_max,SIZE))
c=x+1j*y
z=c.copy()
fractal=np.zeros(z.shape,dtype=np.uint8)+MAX_COLOR
```

```python
for n in range(ITERATION):
    mask=np.abs(z)<=4
    z[mask]=z[mask]**2+c[mask]
    fractal[(fractal==MAX_COLOR)&(-mask)]=(MAX_COLOR-1)*n/ITERATION


plt.subplot(211)
plt.imshow(fractal)
plt.title("mandelbrot")
plt.axis("off")

plt.subplot(212)
mix=np.choose(fractal<lean,[fractal,lean])
plt.imshow(mix)
plt.title("mandelbrot + luna")
plt.axis("off")
plt.show()
```

```python
""" image fuzzy """
NFIGURES = 9
k = numpy.random.random_integers(1, 5, NFIGURES)
a = numpy.random.random_integers(1, 5, NFIGURES)

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']


```

```python
lena = scipy.misc.lena()
matplotlib.pyplot.subplot(211)
matplotlib.pyplot.imshow(lena)
matplotlib.pyplot.axis('off')

matplotlib.pyplot.subplot(212)
blurred = scipy.ndimage.gaussian_filter(lena, sigma=4)

matplotlib.pyplot.imshow(blurred)
plt.axis('off')

theta = numpy.linspace(0, k[0] * numpy.pi, 200)
plt.polar(theta, numpy.sqrt(theta), choice(colors))

for i in xrange(1, NFIGURES):
    theta = numpy.linspace(0, k[i] * numpy.pi, 200)
    matplotlib.pyplot.polar(theta, a[i] * numpy.cos(k[i] * theta), choice(colors))

plt.axis('off')
plt.show()
```

```python
""" mix voice"""
RATE=44100
DTYPE=np.int16
NTONES=50

def generate(freq,amp,dur,phi):
    t=np.linspace(0,dur,dur*RATE)
    data=np.sin(2*np.pi*freq*t+phi)*amp
    return data.astype(DTYPE)

amps=2000.*np.random.random((NTONES,))+200
print amps
durs=0.19*np.random.random((NTONES,))+0.01
print durs
keys=np.random.random_integers(1,88,NTONES)
print keys
freqs=440.*2**((keys-49.)/12.)
print freqs
phis=2*np.pi*np.random.random((NTONES,))
print phis
tone=np.array([],dtype=DTYPE)
print tone
```

```python
for i in xrange(NTONES):
    newtone=generate(freqs[i],amp=amps[i],dur=durs[i],phi=phis[i])
    tone=np.concatenate((tone,newtone))
print tone

plt.plot(np.linspace(0,len(tone)/RATE,len(tone)),tone)
plt.show()
```

```python
""" setup Wave filter """
response = urllib2.urlopen('http://www.thesoundarchive.com/austinpowers/smashingbaby.wav')
print response.info()
WAV_FILE = 'smashingbaby.wav'
filehandle = open(WAV_FILE, 'w')
filehandle.write(response.read())
filehandle.close()
sample_rate, data = scipy.io.wavfile.read(WAV_FILE)
print "Data type", data.dtype, "Shape", data.shape
```

```python
matplotlib.pyplot.subplot(2, 1, 1)
matplotlib.pyplot.title("Original")
matplotlib.pyplot.plot(data)
```

```python

b,a = scipy.signal.iirdesign(wp=0.2, ws=0.1, gstop=60, gpass=1, ftype='butter')


filtered = scipy.signal.lfilter(b, a, data)

matplotlib.pyplot.subplot(2, 1, 2)
matplotlib.pyplot.title("Filtered")
matplotlib.pyplot.plot(filtered)

scipy.io.wavfile.write('filtered.wav', sample_rate, filtered.astype(data.dtype))

matplotlib.pyplot.show()
```
