```python
%pylab inline
from matplotlib.finance import quotes_historical_yahoo as qhy
from datetime import date
```

```python

name="AAPL"
today=date.today()
start=(today.year-1,today.month,today.day)
quotes=qhy("AAPL",start,today)
#print quotes
close=np.array([quote[4] for quote in quotes])
print close
```

```python

logreturn=np.diff(numpy.log(close))
print logreturn
pos=logreturn[logreturn>0]
print pos
```

```python
counts,rets=np.histogram(pos)
rets=rets[:-1]+(rets[1]-rets[0])/2
freqs=1.0/(counts+0.01)
freqs=log(freqs)
print freqs
```

```python
p=np.polyfit(rets,freqs,1)
print p
```

```python
plot(rets,freqs,'o')
plot(rets,p[0]*rets+p[1])
show()
```
