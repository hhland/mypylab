```python
RATE=44100
DTYPE=np.int16
NTONES=50

def generate(freq,amp,dur,phi):
    t=np.linspace(0,dur,dur*RATE)
    data=np.sin(2*np.pi*freq*t+phi)*amp
    return data.astype(DTYPE)
```

```python
amps=2000.*np.random.random((NTONES,))+200
print amps
durs=0.19*np.random.random((NTONES,))+0.01
print durs
keys=np.random.random_integers(1,88,NTONES)
print keys
freqs=440.*2**((keys-49.)/12.)
print freqs
phis=2*np.pi*np.random.random((NTONES,))
print phis
tone=np.array([],dtype=DTYPE)
print tone
```

```python
for i in xrange(NTONES):
    newtone=generate(freqs[i],amp=amps[i],dur=durs[i],phi=phis[i])
    tone=np.concatenate((tone,newtone))
print tone

plt.plot(np.linspace(0,len(tone)/RATE,len(tone)),tone)
plt.show()
    
```
