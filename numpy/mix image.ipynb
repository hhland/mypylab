```python
%pylab inline
import scipy
ITERATION=50
lean=scipy.misc.lena()
SIZE=lean.shape[0]
MAX_COLOR=255
x_min,x_max=-2.5,1
y_min,y_max=-1,1
```

```python
x,y=np.meshgrid(np.linspace(x_min,x_max,SIZE),np.linspace(y_min,y_max,SIZE))
c=x+1j*y
z=c.copy()
fractal=np.zeros(z.shape,dtype=np.uint8)+MAX_COLOR
```

```python
for n in range(ITERATION):
    mask=np.abs(z)<=4
    z[mask]=z[mask]**2+c[mask]
    fractal[(fractal==MAX_COLOR)&(-mask)]=(MAX_COLOR-1)*n/ITERATION
```

```python
plt.subplot(211)
plt.imshow(fractal)
plt.title("mandelbrot")
plt.axis("off")
```

```python
plt.subplot(212)
mix=np.choose(fractal<lean,[fractal,lean])
plt.imshow(mix)
plt.title("mandelbrot + luna")
plt.axis("off")
plt.show()
```
