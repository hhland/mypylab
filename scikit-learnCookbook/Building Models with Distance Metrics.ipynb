```python
%pylab inline
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
hr="-"*120
blobs, classes = make_blobs(500, centers=3)

blobs[0:20,],hr,classes[0:20,]
```

```python
f, ax = plt.subplots(figsize=(7.5, 7.5))
rgb = np.array(['r', 'g', 'b'])
ax.scatter(blobs[:, 0], blobs[:, 1], color=rgb[classes])

ax.set_title("Blobs")
```

```python
kmean=KMeans(n_clusters=3)
kmean.fit(blobs)
```

```python
kmean.cluster_centers_
```

```python
f, ax = plt.subplots(figsize=(7.5, 7.5))
ax.scatter(blobs[:, 0], blobs[:, 1], color=rgb[classes])

ax.scatter(kmean.cluster_centers_[:, 0],
 kmean.cluster_centers_[:, 1], marker='*', s=250,
 color='black', label='Centers')

ax.set_title("Blobs")
ax.legend(loc='best')
```

```python
kmean.labels_[:5],hr,classes[:5],hr,kmean.transform(blobs)[:5]
```

# Optimizing the number of centroids
Centroids are diffiult to interpret, and it can also be very diffiult to determine whether
we have the correct number of centroids. It's important to understand whether your data
is unlabeled or not as this will directly inflence the evaluation measures we can use.

```python
silhouette_samples = metrics.silhouette_samples(blobs,
 kmean.labels_)
silhouette_samples[:20],hr,np.column_stack((classes[:5], silhouette_samples[:5]))
```

```python
f, ax = plt.subplots(figsize=(10, 5))
ax.set_title("Hist of Silhouette Samples")
ax.hist(silhouette_samples)
```
