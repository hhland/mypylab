```python
#http://docs.scipy.org/doc/scipy/reference/tutorial/interpolate.htmlfrom scipy.interpolate import interp1d,griddata as gd
```

```python
x=np.linspace(0,10,10)y=np.cos(-x**2/8.0)f,f2=interp1d(x,y),interp1d(x,y,kind="cubic")
```

```python
xnew=np.linspace(0,10,40)plt.plot(x,y,'o',xnew,f(xnew),'-',xnew,f2(xnew),'--')plt.legend(["data","linear","cubic"],loc="best")plt.show()
```
